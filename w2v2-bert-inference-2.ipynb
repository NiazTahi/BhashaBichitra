```python
from huggingface_hub import login
from kaggle_secrets import UserSecretsClient

user_secrets = UserSecretsClient()
hf_token = user_secrets.get_secret("hf_token")

login(token=hf_token)
```

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition", model="Umong/w2v-bert-2.0-ben10", device=0, batch_size=4)
```

```python
import pandas as pd
```

```python
submit = pd.read_csv("/kaggle/input/ben10/sample_submission.csv")
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
import os

from tqdm.auto import tqdm

sentence = []
audio_dir = "/kaggle/input/ben10/ben10/16_kHz_valid_audio"

for file_id in tqdm(submit["id"]):
    result = pipe(os.path.join(audio_dir, file_id))
    sentence.append(result["text"])
```

```python
submit["sentence"] = sentence
submit.loc[submit["sentence"] == "", "sentence"] = "<>"
submit.to_csv("submission.csv", index=False)
```
